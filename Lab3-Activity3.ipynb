First for this we read the pew data and we read it using read.csv function of R

```r
T1 <- read.csv(file="C://Users//Fnu//Desktop//activity3file.csv",header=TRUE,sep=",")
summary(T1)
str(T1)
```

We first normalize the data and try to make clusters just to see if better results can be obtained

```r
n_sex =  (T1$sex-min(T1$sex))/(max(T1$sex)-min(T1$sex)) 
n_game1 =  (T1$game1-min(T1$game1))/(max(T1$game1)-min(T1$game1)) 
n_age =  (T1$age-min(T1$age))/(max(T1$age)-min(T1$age)) 
n_weight =  (T1$weight-min(T1$weight))/(max(T1$weight)-min(T1$weight)) 
n_emplnw =  (T1$emplnw-min(T1$emplnw))/(max(T1$emplnw)-min(T1$emplnw)) 

t2 <- data.frame(n_sex , n_game1, n_age, n_weight, n_emplnw)
t2
```

We install klaR package to use the kmeans feature #install.packages("klaR",repos="http://cran.us.r-project.org")

```r
library(klaR)
```

We make 3 clusters using kmeans feature and we observe accuracy of 73.1% and we see how every data point is added to each cluster

```r
newk_normalise<- kmeans(t2,3)
newk_normalise
```

Now we take 5 clusters and observe the accuracy of 84.1% which is better than taking 3 clusters

```r
newk_normalise2<- kmeans(t2,5)
newk_normalise2
```

Now we do data cleaning on pew data and tyake few independent variables and amke clusters and observe the accuracy

```r
data.rm.top <- data.frame(T1$sex , T1$age, T1$weight, T1$emplnw, T1$game1)
```

```r
data.rm.top
```

Now we take 5 clusters and observe the accuracy of 90.3% which is better than taking 3 clusters

```r
new_cluster5 <- kmeans(data.rm.top, 5)
new_cluster5
```

Now we take 3 clusters and observe the accuracy of 75.1% which is less than taking 5 clusters

```r
new_cluster3 <- kmeans(data.rm.top, 3)
new_cluster3
```

Now we take different values of K and see which fits best according to the data

```r
rng<-2:20 #K from 2 to 20
tries <-100 #Run the K Means algorithm 100 times
avg.totw.ss <-integer(length(rng)) #Set up an empty vector to hold all of points
for(v in rng){ # For each value of the range variable
 v.totw.ss <-integer(tries) #Set up an empty vector to hold the 100 tries
 for(i in 1:tries){
 k.temp <-kmeans(data.rm.top,centers=v) #Run kmeans
 v.totw.ss[i] <-k.temp$tot.withinss#Store the total withinss
 }
 avg.totw.ss[v-1] <-mean(v.totw.ss) #Average the 100 total withinss
}
plot(rng,avg.totw.ss,type="b", main="Total Within SS by Various K",
 ylab="Average Total Within Sum of Squares",
 xlab="Value of K")
```
