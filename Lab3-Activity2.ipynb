We read the data using the read.csv instruction first

```r
data <- read.csv("C://Users//Fnu//Downloads//lab3Data//german_credit_data1.csv",TRUE)
```

1.Now we use the function str to see the datatypes of various columns of the data so that we can see the columns factored
2.Then we give numeric levels to the factores data columns
3.The data which is not numeric we change into numeric so that we can normalise them

```r
str(data)
#levels(data$Sex)[levels(data$Sex)=="male"] <- 1
#levels(data$Sex)[levels(data$Sex)=="female"] <- 2
levels(data$Sex) <- c(1,2)
levels(data$Housing) <- c(1,2,3)
levels(data$Purpose) <- c(1,2,3,4,5,6,7,8)

data$Saving.accounts <- factor(data$Saving.accounts, levels=c(levels(data$Saving.accounts), "0"))
data$Saving.accounts[is.na(data$Saving.accounts)] = 0
levels(data$Saving.accounts) 
levels(data$Saving.accounts) <- c(0,1,2,3,4)
data$Checking.account <- factor(data$Checking.account, levels=c(levels(data$Checking.account), "0"))
data$Checking.account[is.na(data$Checking.account)] = 0
levels(data$Checking.account) 
levels(data$Checking.account) <- c(0,1,2,3)
data$Saving.accounts <- as.numeric(data$Saving.accounts)
data$Credit.amount <- as.numeric(data$Credit.amount)
data$Age <- as.numeric(data$Age)
data$Duration <- as.numeric(data$Duration)
data$Checking.account <- as.numeric(data$Checking.account)
data$Housing <- as.numeric(data$Housing)
data$Sex <- as.numeric(data$Sex)
data$Purpose <- as.numeric(data$Purpose)
data
```

Now we normalize the data so that the columns whose values have more range comes between 0 and 1

```r
normalized_age =  (data$Age-min(data$Age))/(max(data$Age)-min(data$Age))
normalized_purpose = (data$Purpose-min(data$Purpose))/(max(data$Purpose)-min(data$Purpose))
normalized_sex =  (data$Sex-min(data$Sex))/(max(data$Sex)-min(data$Sex))
normalized_job =  (data$Job-min(data$Job))/(max(data$Job)-min(data$Job))
normalized_credit.amount =  (data$Credit.amount-min(data$Credit.amount))/(max(data$Credit.amount)-min(data$Credit.amount))
normalized_duration =  (data$Duration-min(data$Duration))/(max(data$Duration)-min(data$Duration))
normalised_housing = (data$Housing-min(data$Housing))/(max(data$Housing)-min(data$Housing))
normalised_saving =  (data$Saving.accounts-min(data$Saving.accounts))/(max(data$Saving.accounts)-min(data$Saving.accounts))
noramlised_checking =  (data$Checking.account-min(data$Checking.account))/(max(data$Checking.account)-min(data$Checking.account))
```

Now we tell the number of points and the sampling rate so that we can divide the data into testing and training

# Initially we take the sampling rate as 0.8 to divide the data

```r
n.points <- 1000
sampling.rate <- 0.8
```

```r
num.test <- n.points * (1-sampling.rate)
num.test
```

Now we define how our training data will be and similairly for the test data we will repeat the same process

```r
training <- sample(1:n.points, sampling.rate * n.points,
replace=FALSE)
training
```

We select the columns we need from the data for training and do the same for test also

```r
train <- subset(data[training, ], select = c(normalized_age,normalized_sex,normalized_job,normalised_housing,normalised_saving,noramlised_checking,normalized_credit.amount,normalized_duration,normalized_purpose))
```

```r
testing <- setdiff(1:n.points, training)
testing
```

```r
test <- subset(data[testing, ], select = c(normalized_age,normalized_sex,normalized_job,normalised_housing,normalised_saving,noramlised_checking,normalized_credit.amount,normalized_duration,normalized_purpose))
```

# We define the column on the basis of which we want the result of clustering

```r
cl <- data$Credit.Risks[training]
cl
```

```r
true.labels <- data$Credit.Risks[testing]
true.labels
```

```r
#install.packages("class",repos="http://cran.us.r-project.org")
library(class)
```

Now we take the 20 values of k and show the misclassification rate for all the data and see which k value gives minimum value

```r
for (k in 1:20) {
print(k)
predicted.labels <- knn(train, test, cl, k)
# We're using the R function knn()
num.incorrect.labels <- sum(predicted.labels != true.labels)
misclassification.rate <- num.incorrect.labels /num.test
  print(misclassification.rate)
}
```
